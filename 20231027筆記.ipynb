````python
bmi
````

````python
if bmi < 18.5:
  print("體重過輕")
elif bmi >= 24:
  print("體重太重")
else:
  print("正常體重")
````

````python
if bmi < 18.5:
  print("體重過輕")
elif 18.5 <= bmi < 24:
  print("正常體重")
else:
  print("體重太重")
````

````python
def health01(h, w):
  bmi = w/h**2
  if bmi < 18.5:
    print("體重過輕")
  elif bmi >= 24:
    print("體重太重")
  else:
    print("正常體重")
````

````python
def health02(h, w):
  bmi = w/h**2
  if bmi < 18.5:
    return "體重過輕"
  elif bmi >= 24:
    return "體重太重"
  else:
    return "正常體重"
````

### 例外處理 (Exception Handling)

````python
def add(a, b):
  print(f"{a} + {b} = {a+b}")

def sub(a, b):
  print(f"{a} - {b} = {a-b}")

def mul(a, b):
  print(f"{a} x {b} = {a*b}")

def div(a, b):
  print(f"{a} / {b} = {a/b}")
````

````python
fns = [add, sub, div, mul]
````

````python
for fn in fns:
  fn(1, 2)
````

````python
for fn in fns:
  fn(1, 0)
````

````python
for fn in fns:
  try:
    fn(1, 0)
  except Exception as e:
    print(e.args)
````

````python
for fn in fns:
  try:
    fn(1, 0)
  except Exception as e:
    print('程式出錯了呦')
````

````python
for fn in fns:
  try:
    fn(1, 0)
  except:
    pass
````
