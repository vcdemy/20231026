$ bmi = \frac{體重(公斤)}{身高(公尺)^2} $

|bmi<18.5|體重過輕|
|18.5<=bmi<24|正常體重|
|bmi>=24|體重過重|

```python
h = 1.83
w = 80
```

```python
bmi = w/h**2
```

```python
bmi
```

```python
round(bmi, 2)
```

```python
if bmi < 18.5:
  print("體重過輕")
elif bmi >= 24:
  print("體重太重")
else:
  print("正常體重")
```

```python
if bmi < 18.5:
  print("體重過輕")
elif 18.5 <= bmi < 24:
  print("正常體重")
else:
  print("體重太重")
```

```python
def health01(h, w):
  bmi = w/h**2
  if bmi < 18.5:
    print("體重過輕")
  elif bmi >= 24:
    print("體重太重")
  else:
    print("正常體重")
```

```python
def health02(h, w):
  bmi = w/h**2
  if bmi < 18.5:
    return "體重過輕"
  elif bmi >= 24:
    return "體重太重"
  else:
    return "正常體重"
```

```python
health01(1.83, 80)
```

```python
health02(1.83, 80)
```

```python
x = health01(1.83, 80)
```

```python
y = health02(1.83, 80)
```

### 例外處理 (Exception Handling)

```python
def add(a, b):
  print(f"{a} + {b} = {a+b}")

def sub(a, b):
  print(f"{a} - {b} = {a-b}")

def mul(a, b):
  print(f"{a} x {b} = {a*b}")

def div(a, b):
  print(f"{a} / {b} = {a/b}")
```

```python
add(1, 2)
sub(1, 2)
mul(1, 2)
div(1, 2)
```
